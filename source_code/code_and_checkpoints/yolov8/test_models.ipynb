```python
from ultralytics import YOLO
from pathlib import Path
```

```python
# Get all videos in videos/ directory
videos = [x for x in Path('videos').glob('*.mp4')]
videos
```

```python
detect_model = YOLO("models/best_detect.pt")
detect_model
```

```python
track_model = YOLO("models/best_track.pt")
track_model
```

```python
for video in videos:
    detect_model.predict(video, save_dir="videos/detect")
for video in videos:
    track_model.track(video, save_dir="videos/track")
```
